# Image to Pencil Sketch with Python

```python
import cv2 # opencv is an open source library which provides us with numerous
           # computer vision algorithms
import matplotlib.pyplot as plt
```

## Importing the image from the directory

```python
img = cv2.imread("flower.jpg")
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB), cmap='gray')
plt.show()
```

## Grayscaling the image

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10, 10))
plt.imshow(gray_img, cmap='gray')
plt.title("Grayscaled Image")
plt.show()
```

## Inverting the grayscale image

```python
inv_img = 255 - gray_img
plt.figure(figsize=(10, 10))
plt.imshow(inv_img, cmap='gray')
plt.title("Inverted Image")
plt.show()
```

## Blurring the inverted image

```python
blur = cv2.GaussianBlur(inv_img, (21, 21), 0)
plt.figure(figsize=(10, 10))
plt.imshow(blur, cmap='gray')
plt.title("Blurred Inverted Image")
plt.show()
```

## Sketch of the Image

```python
inv_blur = 255 - blur
p_sk = cv2.divide(gray_img, inv_blur, scale=256)
plt.figure(figsize=(10, 10))
plt.imshow(p_sk, cmap='gray')
plt.title("Pencil Sketch of the Image")
plt.show()
```

## Original Image was:

```python
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB), cmap='gray')
plt.title("Original Image")
plt.show()
```
